```python
#plot distribution of power

plt.figure(figsize=(15,6))
sns.histplot(df['generated_power_kw'],bins=30,kde=True)
plt.title('Distribution of Generated Power (kW)')
plt.xlabel('Generated Power (kW)')
plt.ylabel('frequency')
plt.show()
```

```python
#plot distribution of power

plt.figure(figsize=(10,6))
sns.histplot(df['high_cloud_cover_high_cld_lay'],color='blue',alpha=0.5,label='high_cloud_cover_high_cld_lay',bins=10,kde=True)
sns.histplot(df['medium_cloud_cover_mid_cld_lay'],color='red',alpha=0.5,label='medium_cloud_cover_mid_cld_lay',bins=10,kde=True)
sns.histplot(df['low_cloud_cover_low_cld_lay'],color='green',alpha=0.5,label='low_cloud_cover_low_cld_lay',bins=10,kde=True)
plt.title('Distribution of Angle of Incidence')
plt.xlabel('Cloud cover')
plt.ylabel('frequency')
plt.legend()
plt.show()
```
